# Concat Files 

```python
import pandas as pd
from pandas import DataFrame,Series
import numpy as np
import os
import glob
```

```python
os.chdir("D://Business analytics//Excel class//PowerBI//Power Bi second")
os.getcwd()
```

```python
path = os.getcwd()
csv_files = glob.glob(os.path.join(path,"*.csv"))
csv_files
```

```python
for x in csv_files:
    dframe = pd.read_csv(x)
    print(dframe)
```

```python
csv_list = []
for x in csv_files:
    dframe = pd.read_csv(x)
    dframe.columns=dframe.iloc[1]
    dframe = dframe.iloc[2:]
    csv_list.append(dframe)
csv_list
```

```python
df_final = pd.concat(csv_list)
df_final
```

```python
df_final = df_final.reset_index(drop=True)
df_final.columns
```

```python
df_final["channelKey"].unique()
```

```python
df_final["channelKey"] = df_final["channelKey"].astype(int)
df_final["channelKey"].unique()
```

```python
df_final["StoreKey"] = df_final["StoreKey"].astype(int)
df_final["StoreKey"].unique()
```

```python
df_final["ProductKey"] = df_final["ProductKey"].astype(int)
```

```python
df_final["PromotionKey"] = df_final["PromotionKey"].astype(int)
df_final["GeographyKey"] = df_final["GeographyKey"].astype(int)
```

```python
df_final = df_final.drop(['SalesKey', 'DateKey','PromotionKey','StoreKey'],axis ="columns")
df_final
```

```python
lookup_table = pd.ExcelFile("lookup1.xlsx")
products = pd.read_excel(lookup_table,"Product")
geography = pd.read_excel(lookup_table,"Geography")
channel = pd.read_excel(lookup_table,"Channel")
geography
```

```python
file1 = pd.merge(df_final,products,how="left",left_on = "ProductKey",right_on="ProductKey")
file1 = pd.merge(file1,channel,how="left",left_on="channelKey",right_on="ChannelKey")
file1 = pd.merge(file1,geography,how="left",left_on="GeographyKey",right_on="GeographyKey")
file1
```

```python
file1 = file1.drop(columns =["channelKey","ChannelKey","GeographyKey","ProductKey"])
file1
```

```python
file1.to_csv("final_data.csv")
```
