```python
import pandas as pd
import numpy as np
from pandas import DataFrame,Series
import glob
```

```python
import os
os.getcwd()
```

```python
os.chdir("D://Business analytics//Excel class//Python//files")
os.getcwd()
```

```python
#to access the different files for our case all our csv files so we try to get access and merge that files data
path = os.getcwd()
csv_files = glob.glob(os.path.join(path,"*.csv"))
csv_files
```

```python
#reading different files
for x in csv_files:
    dframe = pd.read_csv(x)
    print(dframe)
```

```python
#first changing the column names in each file as column names are misplaced
#for that purpose we iloc method 
csv_list = []
for x in csv_files:
    dframe = pd.read_csv(x)
    dframe.columns = dframe.iloc[1] #as our original column names appearing at 1 index number row
    dframe = dframe.iloc[2:] #to remove the data before that row as that data is a trash for us
    csv_list.append(dframe)
csv_list
    
```

```python
#now merging or joining of all dataframes in the list
df_final = pd.concat(csv_list)
df_final = df_final.reset_index(drop = True)
df_final
```

```python
null_rows = df_final.isnull().any(axis = 1)
null_values = df_final[null_rows]
null_values #so null values present in our dataframe
```

```python
#usage of unique method -- to give you the all data value in the dataframe
df_final.columns
```

```python
df_final["channelKey"].unique()
```

```python
#as we have duplicate values in respect of string values present in our dataframe so we have to convert that into integer
df_final["channelKey"] = df_final["channelKey"].astype(int)
df_final["channelKey"].unique()
```

```python
#now next target is to find the particular channel names in our dataframe through channel keys just like vlookup
#just droping the particular columns from our dataframe
df_final = df_final.drop(["SalesKey","DateKey","StoreKey","PromotionKey"],axis = "columns")
df_final.columns
```

```python
#access the vlookup excel file-- it is just a reading of the file
vloo_table = pd.ExcelFile("lookup1.xlsx")
channel_data = pd.read_excel(vloo_table,"Channel") #access the sheet through the worksheet name
product_data = pd.read_excel(vloo_table,"Product")
geography_data = pd.read_excel(vloo_table,"Geography")
channel_data
```

```python
#Merging concept -- to merge both the data frames and access the channel names through its keys
dframe_1 = pd.merge(df_final,channel_data, how = "left",left_on = "channelKey",right_on = "ChannelKey")
#now removing the keys from the table
dframe_1 = dframe_1.drop(columns = ["channelKey","ChannelKey"],axis = 1)
dframe_1
```

```python
#next target---same trying to apply for the product keys
df_final["ProductKey"].unique()
```

```python
df_final["ProductKey"] = df_final["ProductKey"].astype(int)
df_final["ProductKey"].unique()
product_data
```

```python
dframe_2 = pd.merge(dframe_1,product_data ,how = "left",left_on = "ProductKey",right_on ="ProductKey")
dframe_2 = dframe_2.drop("ProductKey",axis = 1)
dframe_2
```

```python
#third one same work for Geography names
dframe_2["GeographyKey"] = dframe_2["GeographyKey"].astype(int)
dframe_2["GeographyKey"].unique()
geography_data
```

```python
dframe_3 = pd.merge(dframe_2,geography_data,how = "left",left_on = "GeographyKey",right_on = "GeographyKey")
dframe_3 = dframe_3.drop("GeographyKey",axis = 1)
dframe_3
```

```python
#as we can see there are null values present in our dataframe so we have to replace that one
values = {"ProductName":"No Name","Sub category":"No Sub CAT","Category":"No CAT","ContinentName":"No Info"}
dframe_3 = dframe_3.fillna(value = values)
dframe_3
```

```python
#+convert the our dataframe into excel file
dframe_3.info()
```

```python
dframe_3.to_excel("new_value.xlsx",index= False)
```
