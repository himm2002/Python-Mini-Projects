# GroupBy and Pivot Methods Usage

```python
#testing this functions on small dataset
data = df_final[["Channel","Total Unit Cost"]]
data_1 = data.iloc[:5]
data_1["Section"]= "Section A"
data_1
```

```python
data_2 = data.tail()
data_2["Section"] = "Section B"
data_2
```

```python
concat_data = pd.concat([data_1,data_2])
concat_data = concat_data.reset_index(drop = True)
concat_data
```

Group By Method --- Aggreagting Sum/mean/max/min

```python
#First argument -- columns you want to segregate or groupby
#you can provide more than two column names in this argument
sum_data = concat_data.groupby(["Channel","Section"],as_index = False)["Total Unit Cost"].agg(sum)
sum_data
```

```python
max_data = concat_data.groupby(["Channel","Section"],as_index = False)["Total Unit Cost"].agg(max)
max_data
```

```python
#that is min data for section among the max data in there channels
min_data = max_data.groupby(["Section"],as_index = False)["Total Unit Cost"].agg(min)
min_data
```

# Stack-- Columns to Rows Unstack-- Rows to Columns

```python
#as mean argument not working in the agg function for groupby we have to apply mean() function in group by for series data type
mean_data = concat_data.groupby(["Channel","Section"])["Total Unit Cost"].mean()
mean_data = mean_data.unstack() #we have to unstack the data 
mean_data= mean_data.reset_index()
mean_data = mean_data.rename(columns = {"Section A":"A"})
mean_data
```

```python
concat_data = concat_data.stack()
concat_data
```

# Melt Function

```python
new_data = {"days":[1,2,3,4,5,6,7],"eng":[10,11,12,13,14,15,16],"maths":[20,21,22,23,24,25,26]}
new_df = pd.DataFrame(new_data)
new_df
```

```python
#that will reshape the structure of data columns comes in place of rows with each value of there
df_5 = pd.melt(new_df)
df_5
```

```python
df_5 = pd.melt(new_df,id_vars = "eng") #id var argument gives the column for which you want to make id
df_5 = pd.melt(new_df,id_vars = "days",var_name= "Subjects",value_name ="Marks") #renaming of variable column and value column 
df_5
```

# pivot function

```python
new_data = {"days":[1,2,3,4,5,6,7],
            "student_name":["a","b","c","a","b","c","a"],
            "eng":[10,11,12,13,14,15,16],
            "maths":[20,21,22,23,24,25,26]}
df_work = pd.DataFrame(new_data)
df_work
```

```python
#target-- to make the days index value and columns student name with there values as eng and maths marks
df_7 = df_work.pivot(index = "days",columns = ["student_name"],values = ["eng","maths"])
df_7
#remember one thing pivot only works for non duplicate values as in our df there is no value appearing 2 times
#we cannot do the aggregation function in pivot function \
#now how it will work in duplicate values
new_data ={"days":[1,2,3,1,2,3,1],
            "student_name":["a","b","c","a","b","c","a"],
            "eng":[10,11,12,13,14,15,16],
            "maths":[20,21,22,23,24,25,26]}
df_work = pd.DataFrame(new_data)
df_work
```

```python
df_work_2 = df_work.pivot(index = "days",columns = "student_name",values= ["eng","maths"])
df_work_2 # you can see that will throw an index duplicate values
```

# Pivot table method

```python
#this will work for duplicate values as well we can make the calculations here
df_work_2 = df_work.pivot_table(index = "student_name",columns = "days",values = ["eng","maths"],aggfunc = "mean")
df_work_2
```
