```python
file = pd.read_csv("employees.csv")
file
```

```python
file.columns
```

```python
#through this we are able to know about the null values present in our data
null_values = file.isnull().any(axis=1)
null_data = file[null_values]
null_data
```

```python
#we find null values in the end column exit date...no replacing it with some other values
values = {"Exit Date":"Working"}
file = file.fillna(value = values)
file
```

```python
#filtering --- normal filter through giving conditions
#filtering the age column here for the age > 40 only data show
filtered_data = file[file["Age"] > 40]
filtered_data
```

```python
#sorting--- making in an order to represent the data
#target-- to sort the data on the basis of revenue in ascending order
filtered_data = filtered_data.sort_values(by = "Annual Salary",ascending = False)
filtered_data
```

```python
#target-- now we have to find the mean of annual salary for particular department
filtered_data = filtered_data.rename(columns = {"EEID":"Employee ID"})
filtered_data
```

```python
data_1 = filtered_data[["Department","Annual Salary"]]
data_1 = data_1.reset_index(drop = True)
#conversion command for currency values into normal integer values
data_1["Annual Salary"] = data_1["Annual Salary"].replace("[\$,]","",regex = True).astype(int)
data_1
```

```python
data_1 = data_1.groupby("Department").mean()
data_1 = data_1.reset_index()
data_1
data_1["Annual Salary"] = data_1["Annual Salary"].astype(int)
data_1
```

```python
#program -- to finding which department which annual salary is highest and lowest
salary = []
department = []
for x in data_1["Department"]:
        department.append(x)
for i in data_1["Annual Salary"]:
    salary.append(i)
    
print(salary)
print(department)
#highest value among all the values present in the list
highest_value = max(salary)
#lowest value among all the values present in the list
lowest_value = min(salary)
print(highest_value)
print(lowest_value)
highest_index = salary.index(highest_value)
lowest_index = salary.index(lowest_value)
highest_dept = department[highest_index]
lowest_dept = department[lowest_index]
print("The Highest Annual Salary average among all department is {} with value of {}".format(highest_dept,highest_value))
print("The lowest Annual Salary average among all the department is {} with value of {}".format(lowest_dept,lowest_value))
```
