Pandas-- Dataframe creation 

```python
#To inserting the data into dataframe data type of pandas we are going to use dictionary data type
data_1 = {"Name":["ABC","XYZ","GMW","HIT"],
         "Age": [30,45,67,78],
         "Salary":[90000,50000,40000,30000]}
print(data_1)
```

```python
#Any pandas library work we have to always use run this 3 command
import pandas as pd
import numpy as np #this library for scientific and mathematical calculation
from pandas import Series,DataFrame
```

```python
dframe = pd.DataFrame(data_1)
dframe
#dataframe is the indexed data type with rows and columns
#remember one thing all the pair values should be of equal length otherwise it will through an error
```

```python
#creation of new column in our dataframe with their indexed values
employee_id = ["EMP101","EMP102","EMP103","EMP104"]
dframe["Employee ID"] = employee_id
dframe
```

```python
#doing with the tuple data type
location = ("Chennai","Delhi","Bengaluru","Mysuru")
dframe["Location"] = location
dframe
```

```python
#adding the column at the specfic indexed column number
dframe.insert(3,"Department",["Marketing","HR","Technical","Sales"])
dframe
```

```python
#removing the whole column from data frame-- execute once only otherwise it will throw an error
del dframe["Location"]
dframe
```

```python
#removing the missing data---Concept of data cleaning
data_2 = {"Products":["","","Product C","Product D"],
          "Cost Price":["","",350,600],
          "Selling Price":["","",700,800]}
dframe_2 = pd.DataFrame(data_2)
dframe_2
```

```python
#this method not works for creating the dataframe through dictionary data type as you can see it still not consider that missing value
dframe_2 = dframe_2.dropna(axis = 1)
dframe_2
```
