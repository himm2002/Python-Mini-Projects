Web Scraping Project

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://www.tutorialsfreak.com/python-pandas-tutorial/interview-questions")
page.status_code
```

```python
soup = BeautifulSoup(page.text,"html.parser")
html_page = soup.prettify()
print(html_page)
```

```python
website_title = soup.title
website_title = website_title.text.strip()
print(website_title)
```

```python
questions_data = soup.find("section",class_="Interview-question-section")
print(questions_data)
```

```python
title = questions_data.find("h2",class_= "fs-18 fw-500 label-color-12 pb-2 lh-30")
title = title.text.strip()


questions = []
for i in questions_data.find_all("h3",class_ = "mb-0 fs-18 fw-500 label-color-7 pe-4"):
    i = i.text.strip()
   
    questions.append(i)
answers = []
for i in questions_data.find_all("div",class_= "fs-15 fw-500 label-color-10"):
    i = i.text.strip()
   
    answers.append(i)

print(title)
data_1 = "Q1 " + questions[0] + "\n" +"\n" + answers[0]
print(data_1)
data_2 = "Q2 " + questions[1] + "\n" + "\n" + answers[1]
print(data_2)
```
