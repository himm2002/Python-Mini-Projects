```python
dframe = dframe.dropna(axis=0)
dframe
```

```python
#converting particular column values into some other
dframe['Time'] = dframe['Time'].astype(int)
dframe
```

```python
new_data = dframe[["Location","PopMale","PopFemale"]]
new_data = new_data.reset_index()
new_data
```

```python
new_data = new_data.reset_index()
new_data
```

```python
#grouping the location --- afghanistan and mean of that particular locatio0n
mean_data = new_data.groupby("Location").mean()
mean_data #that is mean of each location here
mean_data = mean_data.reset_index()
mean_data
```

```python
#making new column Population total and make the execution
mean_data["Total Population"] = mean_data["PopMale"] + mean_data["PopFemale"]
mean_data
```
