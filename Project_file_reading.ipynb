```python
file = file.dropna(axis = 0,how="all") #as we have second argument is how that having two values any or all 
#any -- if it finds the null value then it will remove the entire row or column that depends on the axis given
#alln-- if it finds the null value then it will remove only that particular row or column that also depends on the axis
file
```

```python
#removing the whole column with null value in our case it is Value named column
file = file.dropna(axis = 1, how = "all")
file
```

```python
file = file.reset_index(drop = True)
file
```

```python
#conversion of the data type for the particular column values
file.columns
```

```python
file["LocID"] = file["LocID"].astype(int)
file
```

```python
file["Time"] = file["Time"].astype(int)
file
#it showing an error for time column values because there might be some values which is null and cannot able to convert that into integer
```

```python
file["Time"] = file["Time"].fillna(0).astype(int)
file
#in this case Na values got replaced with 0 and then making casting
```

```python
#fetching particular data from our main dataframe like we want only three column data
data_1 = file[["Location","PopFemale","PopMale"]]
data_1 = data_1.reset_index(drop = True)
data_1
```

```python
#groupby method--is like total of the data for the particular column values like what we have done in pivot table
total_data = data_1.groupby("Location").sum()
total_data #this is the sum for each location present in our excel sheet
```

```python
total_data = total_data.reset_index()
total_data
```

```python
#creation of new column with dereiving value from already existing data
total_data["Total Population"] = total_data["PopFemale"] + total_data["PopMale"]

#doing casting to change the decimal values into numbers
total_data["PopFemale"] = total_data["PopFemale"].astype(int)

total_data["PopMale"] = total_data["PopMale"].astype(int)
total_data["Total Population"] = total_data["Total Population"].astype(int)
total_data
```

```python
#target-2 mean of the data for age group
file.columns
```

```python
data_2 = file[["AgeGrp","PopMale","PopFemale","PopTotal"]]
data_2 = data_2.reset_index(drop = True)
data_2
```

```python
mean_data = data_2.groupby("AgeGrp").mean()
mean_data = mean_data.reset_index()
mean_data
```

```python
#now the small program to print the average of the female population
mean_data["PopFemale"] = mean_data["PopFemale"].astype(int)
mean_data["PopMale"] = mean_data["PopMale"].astype(int)


#target-- total average of all the locations for males and compare whether it is grater than female or not
female_list = []
male_list= []
for i in mean_data["PopFemale"]:
    female_list.append(i)
for i in mean_data["PopMale"]:
    male_list.append(i)
print(female_list)
print(male_list)
count = 0
female_total = 0
for x in female_list :
    count = count + 1
    female_total = female_total+ x
female_average = female_total//count
male_total = 0
for x in male_list :
    male_total = male_total + x
male_average = male_total//count
print(male_average)
if female_average > male_average:
    print("Yes The Female population is increasing as {}(female average) is greater than {} (male average)".format(female_average,male_average))
elif female_average == male_average :
    print("Male and female stands equal with {}".format(female_average))
else:
    print("Male population is greater than Female Population with Values {}(male average) greater than {}(female average)".format(male_average,female_average))

  
```
