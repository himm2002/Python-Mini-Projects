# Histogram Plot

```python
import pandas as pd
from pandas import DataFrame,Series
import os
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as st
```

```python
#loading penguins dataset
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
#it will showcase the count of values on the y-axis
sns.displot(penguins["flipper_length_mm"])
plt.show()
```

```python
#Using of bin parameter
sns.displot(penguins["flipper_length_mm"],bins=[170,175,180,185,190,195,200,205,210,215,220,225,230,235],color="g")
plt.show()
```

```python
#Using of Kde paramter that is called kernel density estimator
flights= sns.load_dataset("flights")
flights.head()
```

```python
sns.displot(flights["passengers"],kde=True,bins=[100,150,200,250,300,350,400,450,500,550,600,650],color="b")
plt.title("Distribution of passengers numbers")
plt.show()
```

```python
#using of log_seale as parameter
sns.set_style("darkgrid")
sns.displot(flights["passengers"],log_scale=True,color="g",kde=True)
plt.show()
```

```python
super_market_data = pd.read_excel("supermarket_sales.xlsx")
super_market_data.head()
```

```python
#we are going to use some particular columns not whole dataset
mart_data = super_market_data[["Gender","Unit price","Quantity","Payment","Total","gross income"]]
mart_data
```

```python
#Representation of count of total range values
sns.displot(mart_data["Total"],color="r",bins=[100,200,300,400,500,600,700,800,900,1000])
plt.show()
```

```python
#Same Graph let's see plotting on the y-axis with Gross Income Values
sns.displot(data=mart_data,y="gross income",color="r",bins=[10,20,30,40,50])
plt.show()
```

```python
#Using of bin width parameter
sns.displot(data=mart_data,x="Total",binwidth=30)
plt.show()
```

```python
#better way of giving interval for histogram is using of np.arange
sns.displot(data=mart_data,x="Total",bins= np.arange(0,1100,50),color="y")
plt.show()
```

```python
#now the representation of 0 to 1100 with 50 increasing value interval in graph
sns.set_style("ticks")
plt.figure(figsize=(15,5))
sns.displot(data=mart_data,x="Total",bins= np.arange(0,1100,50),color="b")
plt.xticks(np.arange(0,1100,50))

plt.show()
```

```python
#using of categorical value in hue and stack it using multiple argument
sns.displot(data=mart_data,x="Total",hue="Gender",kde=True,bins=np.arange(0,1100,50))
plt.show()
```

```python
#Now using of nultiple to stack the bars for gender as individual.. because above example is not a good way of representing
sns.displot(data=mart_data,x="Total",hue="Gender",multiple="stack")
plt.show()
```

```python
sns.displot(data=mart_data,x="gross income",hue="Payment",multiple="stack")
plt.show()
```

```python
#Using of parameter element for step or poly(that is similar to something area chart) in histogram
sns.displot(data=mart_data,x="gross income",element="step")
plt.show()
```

```python
sns.displot(data=mart_data,x="gross income",hue="Payment",element="poly")
plt.title("Distribution of Gross income among Payment Modes")
plt.show()
```

```python
#using of stat paramter for histogram for different calculation rather than count it have some limited values
#like probability percent etc.
sns.displot(data=mart_data,x="gross income",hue="Payment",multiple="stack",stat="percent")
```
