```python
import pandas as pd
from pandas import DataFrame,Series
import statistics as st
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_data = sns.load_dataset("tips")
tips_data.head()
```

```python
sns.regplot(data=tips_data,x="total_bill",y="tip")
plt.show()
```

```python
sns.regplot(data=tips_data,x="total_bill",y="tip",color="red",marker="D")
plt.show()
```

```python
#now we are just going to change the color and styling for the line only
sns.regplot(data=tips_data,x="total_bill",y="tip",line_kws=dict(color="black",linestyle="dotted"),marker="o",
             scatter_kws=dict(color="blue"))
```

```python
sns.regplot(data=tips_data,x="total_bill",y="tip",line_kws=dict(color="green",linestyle="dashed"),scatter_kws=dict(
color="orange",s=80,marker="o"))
plt.show()
```

```python
#we have argument ci values you can provide 0-100 or none to show the confidence interval or not
#and if you want the confidence interval then at what percentage 
sns.regplot(data=tips_data,x="total_bill",y="tip",line_kws=dict(color="black"),ci=50)
plt.show()
```

```python
sns.regplot(data=tips_data,x="total_bill",y="tip",line_kws=dict(color="black"),ci=None)
```

```python
mpg_data = sns.load_dataset("mpg")
mpg_data.head()
```

```python
#first we create the linear regression model between weight and accelaration
sns.regplot(data=mpg_data,x="acceleration",y="weight")
plt.show()
```

```python
#in case if we have non linear trends in our dataframe then we increase the value of orders to create flexible line 
#that will reduce the linear regression line
sns.regplot(data=mpg_data,x="acceleration",y="weight",order=4)
plt.show()
```

```python
#using the logx and robust
```

```python
mpg_data.head()
```

```python
sns.regplot(data=mpg_data,x="displacement",y="mpg",logx=True,ci=None)
plt.show()
```

```python
sns.regplot(data=mpg_data,x="horsepower",y="weight")
```

```python
#now creating the same regression plot with roburst that will actually reduce the influence of outliers
sns.regplot(data=mpg_data,x="horsepower",y="weight",robust=True)
plt.show()
```
