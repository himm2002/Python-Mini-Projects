```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
#first plotting basic like years on x -axis and passengers on y-axis
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
sns.lineplot(x="year",y="passengers",data= flights,hue="month",style="month",palette= "husl",markers="o")
plt.title("Monthly passenger Trend Analysis")
plt.show()
```

```python
#in case of yearly distribution
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
sns.lineplot(x="year",y="passengers",data= flights,palette="ch:s=.25,rot=-.25")
plt.title("Yearly passenger Analysis")
plt.show()
```

```python
#for the particular month like July we want the plot for the passengers
july_flights = flights.query("month== 'Jul'")
july_flights
```

```python
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
sns.lineplot(x="year",y="passengers",data= july_flights,palette="dark:#5A9_r")
plt.title("July month passenger Data")
plt.show()
```

```python
#now we are going to first convert the datframe into pivot table that will make easy for plotting

flights_wide = flights.pivot(index="year",columns="month",values="passengers")
flights_wide
```

```python
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
sns.lineplot(flights_wide,dashes= False,markers="o",palette="bright" )

plt.title("Passenger Information")
plt.show()
```

```python
#now changing the plots of the graph like year on y-axis and passengers on x-axis
sns.lineplot(x="passengers",y="year",data=flights,orient="y",dashes=True,markers="o")
plt.title("Testing")
plt.show()
```

# Matplotlib line plotting

```python
#now we are going to see how the line plotting is done through matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

```python
y = np.array([20,50,40,30,50,40,30,70])

plt.plot(y,color="red",linestyle=":",linewidth=2,marker="s")#automatically uses index values as there x-axis
plt.show()
```

```python
#this is some other example of plotting with diamond shaped marker
plt.plot(y,linewidth=1.5,linestyle="--",marker="D",color="green")
plt.show()
```

```python
#double line plotting

y = np.array([20,50,40,30,50,40,30,70])
x = np.array([40,50,20,60,20,40,30,60])
plt.plot(y,color="red",linestyle="--",marker="o")
plt.plot(x,color="green",linestyle=":",marker="o")
plt.legend(["Class1","Class2"])
plt.show()
```

```python
#now using two axis in the plotting
var_1 = np.array(["Item1","Item2","Item3","Item4","Item5"])
var_2 = np.array([10,30,60,30,20])

plt.plot(var_1,var_2,linestyle="-.",linewidth=2,color="black",marker="o")
plt.title("Item Data")
plt.show()
```

# Legend Placing through matplotlib

```python
numbers = np.array([1,8,3,5,2,6,4])
plt.plot(numbers,color="orange",label="First line",marker="o",linestyle="--")
plt.legend(loc=4)
plt.show()
```
