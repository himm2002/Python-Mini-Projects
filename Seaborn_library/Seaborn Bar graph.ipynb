```python
import pandas as pd
from pandas import DataFrame,Series
import numpy as np
import glob
import os
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#using of flights dataset for bar plotting like each month comparison between the years of passengers

data_1 = sns.load_dataset("flights")
data_1.head()
```

```python
sns.barplot(x="month",y="passengers",data=data_1)
plt.title("Monthly passengers comparison")
plt.show()
```

```python
sns.barplot(x="year",y="passengers",data=data_1)
plt.title("Yearly Passenger Comparison")
plt.show()
```

Data Labels Placing

# Count Plot

```python
sns.countplot(x="sex",data=tips_data)
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

This is just like giving the size as per the sex

```python
#now lets see if we do the same plotting with barplot their is the difference
```

```python
ax= sns.barplot(x="sex",y="size",data=tips_data,ci=None)
ax.bar_label(ax.containers[0],fontsize=10)
plt.show()
```

Here You Can see barplot default calculates the mean for the Y-axis value

```python
#through estimator we can change the calculation for the y-axis variable as if we provide the estimator as sum then it will
#produce the same result for what we have done in count but count can be use for other things as well
ax=sns.barplot(x="sex",ci=None,y="size",estimator="sum",data=tips_data)
ax.bar_label(ax.containers[0],fontsize=10)
plt.show()
```

```python
#Now seeing the count of records for each day
ax= sns.countplot(x="day",data=tips_data,palette="flare",saturation=0.5)
ax.bar_label(ax.containers[0],fontsize=10)
plt.show()
```

Estimator Function or Paramter Testing

```python
ax= sns.barplot(x="year",y="passengers",data=data_1,palette="husl",estimator="sum",errorbar=("ci",10),errcolor="b")
ax.bar_label(ax.containers[0],fontsize=9)
plt.title("Number of passengers Travel Across Years")
plt.show()
```

```python
ax = sns.barplot(x="month",y="passengers",data=data_1,palette="husl",estimator="max",ci=None)
ax.bar_label(ax.containers[0],fontsize=10)
plt.title("Maximum Passenger for each month across years")
plt.show()
```

# Using of Catplot to Represent multiple Bar Charts for Different Categories in same Bar

```python
sns.catplot(data=tips_data,kind="bar",x="sex",y="total_bill",col="time",height=4,aspect=0.5,estimator="sum")
#remember we cannnot use the bar_labels in facet grid or cat plot visuals
plt.show()
```

```python
sns.catplot(data=data_1,kind="bar",x="month",y="passengers",col="year",height=4,aspect=0.6,estimator="sum",palette="husl")
plt.show()
```

```python
ax = sns.catplot(data=data_1,kind="bar",x="passengers",y="month",orient="h",estimator="min",palette="husl",saturation=4,
                ci=50)
ax.bar_label(ax.containers[0],fontsize=10)
plt.show()
```
