# Line Plotting

```python
var_1 = [1,2,3,4,5,6,7]
var_2 = [2,3,4,1,5,6,7]

#in case of seaborn you cannot plot until you convert your dataset into pandas dataframe

#below error is the proof that seaborn not works with only two variables

#sns.lineplot(x="var_1",y="var_2")
```

Matplotlib library Plotting

```python
#first we are going to create line plot through matplotlib

plt.plot(var_1,var_2) #first argument for x-axis and second argument for y-axis
plt.show()
```

```python
#for seaborn we are going to create dataframe

x_1 = pd.DataFrame({"var_1":var_1,"var_2":var_2})
x_1
```

```python
sns.lineplot(x="var_1",y="var_2",data= x_1)
plt.show()
```

```python
#in case if you don't want the labels provided in the dataframe as your axis names you can direcly access the values
var_1 =  [1,2,3,4,5,6,7]
var_2 = [2,3,4,1,5,6,7]

sns.lineplot(x=var_1,y=var_2)
plt.show() #now this will look like similar as matplotlib graph
```

# Line Plotting on big dataset

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.lineplot(x= "bill_length_mm",y= "bill_depth_mm",data= penguins)
plt.show()
```

```python
#using of hue argument
sns.lineplot(x= "bill_length_mm",y= "bill_depth_mm",data= penguins,hue= "sex",style="sex")
plt.show()
```

```python
#now using palette argument
sns.lineplot(x= "bill_length_mm",y= "bill_depth_mm",data= penguins,hue= "sex",style="sex",palette="flare")
plt.show()
```

```python
#Using of markers with now reducing the data size
data_1 = penguins.head(50)
sns.lineplot(x= "bill_length_mm",y= "bill_depth_mm",data=data_1,hue= "sex",
             style="sex",palette="dark:#5A9_r",markers= ["o",">"])
plt.grid()
plt.title("First Attempt")
plt.show()
```

```python
#using of legends or not 
sns.lineplot(x= "bill_length_mm",y= "bill_depth_mm",data=data_1,hue= "sex",
             style="sex",palette="dark:#5A9_r",markers= ["o",">"],dashes=True,legend= False)
plt.show()
```

# Scatter Plot

```python
tips_data = sns.load_dataset("tips")
tips_data
```

```python
#first scatterplot graph through seaborn library
sns.scatterplot(x="tip",y="total_bill",data=tips_data)
plt.show()
```

```python
#we can make the segregration for our data through any column like sex
sns.scatterplot(x="tip",y="total_bill",data=tips_data,hue="sex",)
plt.show()
```

```python
#or we can make the segregration through days colummn with sex column
sns.scatterplot(x="tip",y="total_bill",data=tips_data,hue="day",style="sex",palette= "tab10")
plt.show()
```

```python
#now changing the axis and seeing the result
sns.scatterplot(x="total_bill",y="tip",data=tips_data,hue="day",palette= "tab10")
plt.show()
```

```python
#now using of size argument in scatter plotting
sns.scatterplot(x="total_bill",y="tip",data=tips_data,hue="sex",style="sex",size="sex",sizes=(100,40),palette= "Dark2_r")
```

We can easily watch the female plots that's the use of size and sizes in the scatter plotting

```python
#now lets see how we can create the markers and add it in our plotting
user_marker = {"Male":"o","Female":"s"}
sns.scatterplot(x="total_bill",y="tip",data=tips_data,hue="sex",style="sex",size="sex",sizes=(80,40),palette= "magma_r",
               markers=user_marker)
plt.title("Tips Analysis")
plt.show()
```
