Remember one thing in scatter plotting--- Both Varibles should be in numeric values 

```python
sns.scatterplot(x="Spending_USD",y="Life_Expectancy",data =health_exp,hue="Country")
plt.title("Spending and Expectancy relationship")
plt.show()
```

```python
#if we put the same variable in style it will produce the markers for you and we can provide our own markers in it
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
sns.scatterplot(x="Spending_USD",y ="Life_Expectancy",data= health_exp,hue="Country",style="Country")
plt.show()
```

```python
#now assigning the hue and style different let see the result
sns.scatterplot(x="Spending_USD",y ="Life_Expectancy",data= health_exp,hue="Year",style="Country")
plt.show()
```

```python
#evaluating the tip rate from total bill and tip 
tips_data = sns.load_dataset("tips")
tips_data
```

```python
tip_rate = tips_data.eval("tip/total_bill").rename("tip_rate")
sns.scatterplot(x="total_bill",y="tip",data= tips_data,hue=tip_rate,size=tip_rate,sizes=(40,100))
plt.show()
```

```python
#for normalizing the colors we can pass hue_norm or plt.colors.normalize
sns.scatterplot(x="total_bill",y="tip",data= tips_data,hue="size",size="size",sizes=(40,100),hue_norm=(0,7))
```

```python
dframe_1 = pd.read_excel("mart_train_sample.xlsx")
dframe_1
```

```python
#first scatter plotting for varaibles sales and MRP
sns.scatterplot(data=dframe_1,x="Item_MRP",y="Sales")
plt.show()
```

Looks like that we have not a good coorelation between this two variables thats why it is unevenly spread

```python
#now we are going to use hue argument for grouping basis of outlet
sns.scatterplot(data=dframe_1,x="Item_MRP",y="Sales",hue="Outlet_Size")
plt.show()
```

```python
shapes = {"Small":"o","Medium":"s","High":"D"}
sns.scatterplot(data=dframe_1,x="Item_MRP",y="Sales",hue="Outlet_Size",style="Outlet_Size",markers= shapes)
plt.show()
```

```python
#now lets making grouping for both outlet size and outlet location
sns.scatterplot(data=dframe_1,x="Item_MRP",y="Sales",hue="Outlet_Location_Type",style="Outlet_Size",markers= shapes)
plt.xlabel("Product MRP")
plt.ylabel("Product Sales")
plt.show()
```

```python
#Take an example like this is dataframe one value
x_1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y_1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
#Take an example like this is dataframe two values
x_2 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y_2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])



plt.scatter(x_1,y_1,marker="o",s=100)
plt.scatter(x_2,y_2,marker="D",s=50)
plt.legend(["Day1","Day2"])
plt.show()
```

```python
#now we are also going to change the size for each dot for more representation
x_1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y_1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
#Take an example like this is dataframe two values
x_2 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y_2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])

#sizes
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x_1,y_1,marker="o",s=(20,50,100,200,500,1000,60,90,10,300,600,800,75))
plt.scatter(x_2,y_2,marker="D",s=(60))
plt.legend(["Day_1","Day_2"])
plt.xlabel("Quantity Sold")
plt.ylabel("Sales Value")
plt.show()
```
